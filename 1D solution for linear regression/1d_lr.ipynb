```python
import csv
import numpy as np
import matplotlib.pyplot as plt


#Load the data, using arrays
X=[]
Y=[]

for line in open('data_1d.csv'):
    x, y = line.split(',')
    X.append(float(x))
    Y.append(float(y))
```

```python

#Now lets use numpy arrays

X = np.array(X)
Y = np.array(Y)

#Let's plot the data!

plt.scatter(X,Y)
plt.show()
```

```python

#Now let's apply the equations that we learned to find the best a & b

denominator = X.dot(X)-X.mean()*X.sum()

a=(X.dot(Y)- Y.mean()*X.sum())/denominator

b= (Y.mean()*X.dot(X)-X.mean()*(X.dot(Y)))/denominator

#Calculate predicted Y (The linear regression)

Yi=a*X+b
print(a)
print(b)

plt.scatter(X,Y)
plt.plot(X,Yi)
plt.show()
```

```python
denominator = X.dot(X)/len(X)-(X.mean())**2

a= (X.dot(Y)/len(X) - X.mean()*Y.mean())/denominator

b= ( Y.mean()*X.dot(X)/len(X) - X.mean()* X.dot(Y)/len(X))/denominator

#Calculate predicted Y (The linear regression)

Yi=a*X+b
print(a)
print(b)
plt.scatter(X,Y)
plt.plot(X,Yi)
plt.show()
```

$R^2$ in code

```python
#Let's focus on the denominator

d1= Y - Yi 
d2 = Y - Y.mean()

r2= 1 - d1.dot(d1) / d2.dot(d2)

print(f"The R-squared is: {r2}")
```
