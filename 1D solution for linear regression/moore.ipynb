```python
import re
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt


#Load the data, using arrays
X=[]
Y=[]

non_decimal = re.compile(r'[^\d]+')

for line in open('moore.csv'):
    r = line.split('\t')
    x = int(non_decimal.sub('', r[2].split('[')[0]))
    y = int(non_decimal.sub('', r[1].split('[')[0]))
    X.append(x)
    Y.append(y)
```

```python

#Now lets use numpy arrays

X = np.array(X)
Y = np.array(Y)

#Let's plot the data!

plt.scatter(X,Y)
plt.show()
```

```python
#Y axis in log scale
Y=np.log(Y)
```

```python
#Graph the graph with the new scale!
plt.scatter(X,Y)
plt.show()
```

```python
#Now let's make the linear regresion


denominator = X.dot(X)-X.mean()*X.sum()

a=(X.dot(Y)- Y.mean()*X.sum())/denominator

b= (Y.mean()*X.dot(X)-X.mean()*(X.dot(Y)))/denominator

#Calculate predicted Y (The linear regression)

Yi=a*X+b
print(a)
print(b)

plt.scatter(X,Y)
plt.plot(X,Yi)
plt.show()
```

```python
#The value of the slope gives us the informaion of the number of transistor

time_d=np.log(2)/a
print(time_d) #This gives us almost 2 years
```

```python
#Now let's check the $R^2$


d1= Y - Yi 
d2 = Y - Y.mean()

r2= 1 - d1.dot(d1) / d2.dot(d2)

print(f"The R-squared is: {r2}")
```
