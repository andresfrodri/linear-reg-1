```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50
D=50
#"Fat" matrix

X = (np.random.random((N,D))-0.5)*10 #random number distributed arround 0 from minus 5 to 5

true_w = np.array([1, 0.5, -0.5] + [0]*(D-3))

Y = X.dot(true_w) + np.random.randn(N)*0.5
```

```python
costs = []

#Let's start with the gradient descent
w = np.random.randn(D) / np.sqrt(D)

learning_rate = 0.001

l1 = 10.0

for step in range(500):
    Y_pred = X.dot(w)
    delta = Y_pred - Y 
    w = w - learning_rate*(X.T.dot(delta) + l1*np.sign(w))

    mse = delta.dot(delta) / N

    costs.append(mse)

plt.plot(costs)
plt.show()
```

```python
print(f"final w: {w}")

plt.plot(true_w, label= "True W")
plt.plot(w, label="W MAP")
plt.legend
plt.show()
```
