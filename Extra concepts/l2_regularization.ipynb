```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Let's generate some data

N=50
X = np.linspace(0, 10, N)
Y = 0.5*X +np.random.randn(N) #Random noise

#Manually set outliers

Y[-1] += 30
Y[-2] += 30

plt.scatter(X, Y)
plt.show()
```

```python
#Add the bias term

X = np.vstack([np.ones(N), X]).T


#Calculate the maximun likelihood solution

w_ml = np.linalg.solve(X.T.dot(X), X.T.dot(Y))

Ypred_ml = X.dot(w_ml)


#Let's see the result of the maximun likelihood
plt.scatter(X[:,1],Y)
plt.plot(X[:,1], Ypred_ml)
plt.show()
```

```python
#Now let's see how this changes with L2 regularization 
#penalizing the large weights 

l2 = 1000.0
w_MAP = np.linalg.solve(l2*np.eye(2) + X.T.dot(X), X.T.dot(Y))


Ypred_MAP = X.dot(w_MAP)

plt.scatter(X[:,1],Y)
plt.plot(X[:,1], Ypred_ml, label = "Maximum likelihood")
plt.plot(X[:,1], Ypred_MAP, label = "Maximum aposteriori(MAP)")


plt.legend()
plt.show()
```

In this example we see how to "fight" the outliers using L2 regularization
