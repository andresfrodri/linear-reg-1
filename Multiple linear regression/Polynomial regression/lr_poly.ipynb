```python
import csv
import numpy as np
import matplotlib.pyplot as plt

```

```python
#Load the data as arrays
X=[]
Y=[]

for line in open('data_poly.csv'):
    x, y = line.split(',')
    x =float(x)
    X.append([1, x, x*x])
    Y.append(float(y))
```

```python
#Now lets use numpy arrays

X = np.array(X)
Y = np.array(Y)
```

```python
#Let's plot the data!

plt.scatter(X[:,1],Y)
plt.show()
```

```python
#Calculate the weights

w = np.linalg.solve(np.dot(X.T,X), np.dot(X.T, Y))

Y_pred = np.dot(X,w)
```

```python
#Plot it all together

plt.scatter(X[:,1],Y)
plt.plot(sorted(X[:,1]),sorted(Y_pred))

plt.show()
```

```python
#Let's calculated the R-squared
d1 = Y - Y_pred

d2 = Y - Y.mean()

r2= 1- d1.dot(d1)/d2.dot(d2)

print(f"The r-squared is {r2}")
```
