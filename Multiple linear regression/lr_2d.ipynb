```python
import csv
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt 
```

Load the data

```python
X = []
Y = []
for line in open('data_2d.csv'):
    x1, x2, y = line.split(',')
    X.append([1,float(x1),float(x2)])
    Y.append(float(y))

    
```

```python
#Now lets use numpy arrays

X = np.array(X)
Y = np.array(Y)


```

```python
#Let's plot the data!


fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X[:,1],X[:,2],1)
plt.show()
```

```python
#Calculate the weights 
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T,Y))

Y_pred = np.dot(X,w)

```

```python
d1 = Y - Y_pred

d2 = Y - Y.mean()

r2= 1- d1.dot(d1)/d2.dot(d2)

print(f"The r-squared is {r2}")
```
