```python
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('scrap price.csv')

df.head()
```

```python
#We will select a few of the 26 columns enginesize, carbody, horsepower, price

Y=np.array(df[['price']])
X1=np.array(df[['enginesize']])
X2 = np.array(df[['horsepower']])
```

```python
X=np.concatenate([X1, X2],axis=1)
print(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X[:,0],X[:,1],1)
plt.show()
```

```python
#Calculate the weights 
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T,Y))

Y_pred = np.dot(X,w)
```

```python
d1 = Y - Y_pred

d2 = Y - Y.mean()

r2= 1- d1[:,0].dot(d1[:,0])/d2[:,0].dot(d2[:,0])

print(f"The r-squared is {r2}")
```
